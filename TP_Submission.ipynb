# 📝 Formulaire de Soumission de TP

Remplissez le formulaire ci-dessous pour envoyer votre travail pratique.

```python
# Installation des dépendances nécessaires
import sys
!{sys.executable} -m pip install -q python-dotenv ipywidgets
import smtplib
from email.message import EmailMessage
import os
from dotenv import load_dotenv
from datetime import datetime
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import json

# Charger les variables d'environnement
load_dotenv()

email_sender = os.getenv("EMAIL_SENDER")
email_password = os.getenv("EMAIL_PASSWORD")
email_receiver = os.getenv("EMAIL_RECEIVER")

# Fonction d'envoi d'email
def envoyer_email(nom, prenom, tp_type):
    """
    Envoie un email avec les informations du TP
    """
    try:
        date_envoi = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        
        msg = EmailMessage()
        msg['Subject'] = f"TP Soumission - {tp_type}"
        msg['From'] = email_sender
        msg['To'] = email_receiver
        msg.set_content(
            f"Bonjour Monsieur,\n\n"
            f"Veuillez trouver ci-joint mon Travail Pratique (TP).\n\n"
            f"Informations de l'étudiant :\n"
            f"Nom : {nom}\n"
            f"Prénom : {prenom}\n"
            f"Type de TP : {tp_type}\n\n"
            f"Date d'envoi : {date_envoi}\n\n"
            f"Cordialement."
        )
        
        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as smtp:
            smtp.login(email_sender, email_password)
            smtp.send_message(msg)
        
        return True, "Email envoyé avec succès!"
    except Exception as e:
        return False, f"Erreur lors de l'envoi: {str(e)}"

# Créer le formulaire
print("\n🎓 FORMULAIRE DE SOUMISSION DE TP\n" + "="*50 + "\n")

# Champs d'entrée
nom_input = widgets.Text(
    placeholder='Entrez votre nom',
    description='Nom:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='400px')
)

prenom_input = widgets.Text(
    placeholder='Entrez votre prénom',
    description='Prénom:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='400px')
)

tp_dropdown = widgets.Dropdown(
    options=['Ex1_TP3', 'Ex2_TP3', 'Ex3_TP3', 
             'Ex4_TP3', 'Ex5_TP3', 'Ex6_TP3', 
             'Ex7_TP3'],
    description='Sélectionnez le TP:',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='400px')
)

# Bouton d'envoi
envoyer_btn = widgets.Button(
    description='📤 Envoyer',
    button_style='success',
    tooltip='Cliquez pour envoyer votre TP',
    layout=widgets.Layout(width='200px', height='40px')
)

# Message de statut
output_message = widgets.Output()

# Fonction de gestion du clic sur le bouton
def on_envoyer_click(b):
    with output_message:
        clear_output()
        
        # Validation des champs
        if not nom_input.value.strip():
            print("❌ Erreur: Le nom est obligatoire!")
            return
        
        if not prenom_input.value.strip():
            print("❌ Erreur: Le prénom est obligatoire!")
            return
        
        if not tp_dropdown.value:
            print("❌ Erreur: Veuillez sélectionner un TP!")
            return
        
        # Envoyer l'email
        succes, message = envoyer_email(
            nom_input.value.strip(),
            prenom_input.value.strip(),
            tp_dropdown.value
        )
        
        if succes:
            print(f"✅ {message}")
            print(f"\n📋 Résumé de votre soumission:")
            print(f"   Nom: {nom_input.value}")
            print(f"   Prénom: {prenom_input.value}")
            print(f"   TP: {tp_dropdown.value}")
            print(f"   Date/Heure: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
            
            # Réinitialiser les champs
            nom_input.value = ''
            prenom_input.value = ''
        else:
            print(f"❌ {message}")

envoyer_btn.on_click(on_envoyer_click)

# Afficher le formulaire
form = widgets.VBox([
    nom_input,
    prenom_input,
    tp_dropdown,
    widgets.HTML("<br>"),
    envoyer_btn,
    output_message
])

display(form)
```
