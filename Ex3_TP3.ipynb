### QUESTION 3 : Méthode du Gradient Conjugué pour f(x,y)

```python
# ============================================================
#
# Algorithme du Gradient Conjugué (Fletcher-Reeves) :
#   Initialisation : x₀ donné, r₀ = b - A·x₀, d₀ = r₀
#   Pour k = 0, 1, 2, ... :
#     αₖ = (rₖᵀ · rₖ) / (dₖᵀ · A · dₖ)
#     xₖ₊₁ = xₖ + αₖ · dₖ
#     rₖ₊₁ = rₖ - αₖ · A · dₖ
#     βₖ = (rₖ₊₁ᵀ · rₖ₊₁) / (rₖᵀ · rₖ)
#     dₖ₊₁ = rₖ₊₁ + βₖ · dₖ
# ============================================================

def gradient_conjugue(A, b, x0, tol=1e-10, max_iter=100, verbose=True):
    """
    Implémentation de la méthode du Gradient Conjugué.
    
    Paramètres :
    -----------
    A : ndarray
        Matrice symétrique définie positive de taille (n, n)
    b : ndarray
        Vecteur de taille n
    x0 : ndarray
        Point initial de taille n
    tol : float
        Tolérance pour le critère d'arrêt (norme du résidu)
    max_iter : int
        Nombre maximum d'itérations
    verbose : bool
        Si True, affiche les détails de chaque itération
    
    Retourne :
    --------
    x : ndarray
        Solution approximée
    history : dict
        Historique des itérations (x, r, d, alpha, beta)
    """
    # TODO : Complétez le code suivant.
    x = ......
    r = ......        # Résidu initial : r₀ = b - A·x₀
    d = ......         # Direction initiale : d₀ = r₀
    
    # Historique pour traçage et analyse
    history = {
        'x': [x.copy()],
        'r': [r.copy()],
        'd': [d.copy()],
        'alpha': [],
        'beta': [],
        'residu_norm': [np.linalg.norm(r)]
    }
    
    if verbose:
        print("=== QUESTION 3 : Gradient Conjugué (mode verbose) ===")
        print(f"\nInitialisation :")
        print(f"  x₀ = {x}")
        print(f"  r₀ = b - A·x₀ = {r}")
        print(f"  d₀ = r₀ = {d}")
        print(f"  ||r₀|| = {np.linalg.norm(r):.6f}")
        print("\n" + "="*60)
    
    for k in range(max_iter):
        # Calcul du pas optimal αₖ par minimisation le long de dₖ
        # αₖ = (rₖᵀ · rₖ) / (dₖᵀ · A · dₖ)
        r_dot_r = np.dot(r, r)
        dAd = np.dot(d, A @ d)
        alpha = r_dot_r / dAd
        
        # Mise à jour de la solution
        x_new = x + alpha * d
        
        # Mise à jour du résidu
        r_new = r - alpha * (A @ d)
        
        # Calcul de la norme du nouveau résidu
        residu_norm = np.linalg.norm(r_new)
        
        if verbose:
            print(f"\nItération {k+1} :")
            print(f"  α_{k} = (r_{k}ᵀ·r_{k}) / (d_{k}ᵀ·A·d_{k}) = {r_dot_r:.6f} / {dAd:.6f} = {alpha:.6f}")
            print(f"  x_{k+1} = x_{k} + α_{k}·d_{k} = {x_new}")
            print(f"  r_{k+1} = r_{k} - α_{k}·A·d_{k} = {r_new}")
            print(f"  ||r_{k+1}|| = {residu_norm:.10f}")
        
        # Stockage dans l'historique
        history['x'].append(x_new.copy())
        history['r'].append(r_new.copy())
        history['alpha'].append(alpha)
        history['residu_norm'].append(residu_norm)
        
        # Critère d'arrêt : convergence atteinte
        if residu_norm < tol:
            if verbose:
                print(f"\n✓ Convergence atteinte après {k+1} itération(s)!")
            break
        
        # Calcul du coefficient βₖ (Fletcher-Reeves)
        # βₖ = (rₖ₊₁ᵀ · rₖ₊₁) / (rₖᵀ · rₖ)
        r_new_dot_r_new = np.dot(r_new, r_new)
        beta = r_new_dot_r_new / r_dot_r
        
        # Nouvelle direction de recherche
        # dₖ₊₁ = rₖ₊₁ + βₖ · dₖ
        d_new = r_new + beta * d
        
        if verbose:
            print(f"  β_{k} = (r_{k+1}ᵀ·r_{k+1}) / (r_{k}ᵀ·r_{k}) = {r_new_dot_r_new:.6f} / {r_dot_r:.6f} = {beta:.6f}")
            print(f"  d_{k+1} = r_{k+1} + β_{k}·d_{k} = {d_new}")
        
        history['d'].append(d_new.copy())
        history['beta'].append(beta)
        
        # Préparation pour l'itération suivante
        # TODO : Complétez le code suivant.
        x = ....
        r = ....
        d = ....
    # TODO : Complétez le code suivant.
    return ...., history


# --- Exécution pour f(x,y) ---
x0_f = np.array([0.0, 0.0])  # Point initial donné

x_opt_f, hist_f = gradient_conjugue(A_f, b_f, x0_f, verbose=True)

print(f"\nSolution numérique : x* = {x_opt_f}")
```

### QUESTION 5 : Nombre d'itérations théoriques pour convergence exacte

```python
# ============================================================
# ============================================================
#
# Théorème fondamental du Gradient Conjugué :
# Pour une fonction quadratique à n variables, la méthode du
# gradient conjugué converge en AU PLUS n itérations (en arithmétique
# exacte, sans erreurs d'arrondi).
#
# Pour f(x,y) : n = 2 variables (x et y)
# → Convergence théorique en AU PLUS 2 itérations
#
# En pratique (arithmétique flottante), on peut avoir besoin de
# quelques itérations supplémentaires à cause des erreurs d'arrondi.
# ============================================================

print("=== QUESTION 5 : Itérations théoriques ===")
print()

n_variables = A_f.shape[0]  # Dimension de l'espace (2 pour f)

print(f"Dimension de l'espace : n = {n_variables}")
print(f"Nombre d'itérations effectuées : {len(hist_f['x']) - 1}")
print()
print("Théorème : Pour une fonction quadratique à n variables,")
print(f"le gradient conjugué converge en AU PLUS n = {n_variables} itérations")
print("(en arithmétique exacte, sans erreurs d'arrondi).")
print()
print(f"→ Pour f(x,y) (n={n_variables}), convergence théorique en ≤ {n_variables} itérations.")

# Vérification du nombre d'itérations effectives
n_iter_effectif = len(hist_f['x']) - 1
print(f"\nItérations effectives : {n_iter_effectif}")
print(f"Théorique maximum   : {n_variables}")

if n_iter_effectif <= n_variables:
    print(f"\n✓ Résultat conforme au théorème ! ({n_iter_effectif} ≤ {n_variables})")
else:
    print(f"\n⚠ Plus d'itérations que prévu (erreurs d'arrondi flottant)")
```

## Fonction g(x,y,z) = x² + 2y² + 3z² + 2xy - z

### QUESTION 1 : Forme quadratique de g(x,y,z) = x² + 2y² + 3z² + 2xy - z

```python
print("=== QUESTION 1 : Forme quadratique de g(x,y,z) ===")
print()

# Définition symbolique
x_s, y_s, z_s = sp.symbols('x y z', real=True)
g_sym = x_s**2 + 2*y_s**2 + 3*z_s**2 + 2*x_s*y_s - z_s

print(f"g(x,y,z) = {g_sym}")
print()
# TODO : Complétez le code suivant.
# Matrice A (symétrique)
A_g = np.array([[...., 2, 0],
                [2, 4, ....],
                [0, ....., 6]], dtype=float)
# TODO : Complétez le code suivant.
# Vecteur b
b_g = np.array([0, ...., 1], dtype=float)

# Scalaire c
c_g = 0.0

print("Matrice A :")
print(A_g)
print()
print(f"Vecteur b : {b_g}")
print(f"Scalaire c : {c_g}")
print()

# Vérification avec SymPy
X_g = sp.Matrix([x_s, y_s, z_s])
# TODO : Complétez le code suivant.
A_g_sym = sp.Matrix([[2, ...., 0], [2, ...., 0], [0, ....., 6]])
b_g_sym = sp.Matrix([0, 0, 1])

quad_form_g = sp.Rational(1,2) * (X_g.T * A_g_sym * X_g)[0] - (b_g_sym.T * X_g)[0]
print(f"Vérification : ½ xᵀAx - bᵀx = {sp.expand(quad_form_g)}")
print(f"Original     : g(x,y,z) = {sp.expand(g_sym)}")
print(f"Match : {sp.expand(quad_form_g) == sp.expand(g_sym)}")
```

### QUESTION 2 : Points critiques de g(x,y,z) (résolution analytique)

```python
# On résout ∇g(x,y,z) = 0
#
# g(x,y,z) = x² + 2y² + 3z² + 2xy - z
# ∂g/∂x = 2x + 2y = 0   →  x + y = 0
# ∂g/∂y = 4y + 2x = 0   →  x + 2y = 0
# ∂g/∂z = 6z - 1 = 0    →  z = 1/6
#
# De x + y = 0 et x + 2y = 0 :
#   y = 0, x = 0
#
# Point critique : (0, 0, 1/6)
# ============================================================

print("=== QUESTION 2 : Points critiques de g(x,y,z) ===")
print()

# Calcul du gradient symbolique
grad_g = [sp.diff(g_sym, x_s), sp.diff(g_sym, y_s), sp.diff(g_sym, z_s)]
print("Gradient ∇g(x,y,z) :")
print(f"  ∂g/∂x = {grad_g[0]}")
print(f"  ∂g/∂y = {grad_g[1]}")
print(f"  ∂g/∂z = {grad_g[2]}")
print()

# Résolution du système ∇g = 0
critical_points_g = sp.solve(grad_g, [x_s, y_s, z_s])
print(f"Points critiques (solution analytique) : {critical_points_g}")

# Extraction du point critique
x_star_g = np.array([
    float(critical_points_g[x_s]),
    float(critical_points_g[y_s]),
    float(critical_points_g[z_s])
])
print(f"x* = {x_star_g}")

# Vérification : A x* = b
print()
print("Vérification : A·x* = b ?")
print(f"A·x* = {A_g @ x_star_g}")
print(f"b    = {b_g}")
print(f"Match : {np.allclose(A_g @ x_star_g, b_g)}")
```
